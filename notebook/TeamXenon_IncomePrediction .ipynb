<h3>Project Title:</h3>

<h3>Introduction:</h3>
<strong>Income inequality</strong> is a significant socio-economic challenge faced by developing nations worldwide. Addressing this issue requires a deep understanding of income distribution patterns and factors influencing them. This project aims to leverage machine learning techniques to predict an individual's income level based on various demographic and socio-economic attributes. By accurately predicting income levels, policymakers can better understand and manage income inequality, thereby fostering more equitable economic growth.

<h3>Business Understanding:</h3>
<strong>Understanding income distribution patterns</strong> is crucial for policymakers, economists, and social scientists to address income inequality effectively. By predicting income levels, policymakers can:

<ul>
  <li>Identify vulnerable populations: By analyzing factors influencing income levels, policymakers can identify demographics or socio-economic groups at higher risk of poverty or marginalization.</li>
  <li>Target interventions: Predictive models can help target social welfare programs, educational initiatives, or economic policies towards populations most in need.</li>
  <li>Monitor progress: Regular predictions of income levels can serve as a monitoring tool to evaluate the effectiveness of poverty alleviation programs or economic reforms.</li>
</ul>

### Percentage-wise Analysis of missing values

```python
# Define a function to print the missing values in the datasets
datasets = {'train': train_df, 'test': test_df}

# Define a function to print the missing values in the datasets
def show_missing_values(datasets):
    for name, data in datasets.items():
        title = f"Missing values in the {name.capitalize()} dataset:"
        underline = '*' * len(title)
        print(title)
        print(underline)

        # Calculate the sum of missing values for each column
        missing_values = data.isnull().sum()

        # Exclude columns with missing value percentages below or equal to 50%
        missing_values = missing_values[missing_values / len(data) * 100 > 50]

        # Check if there are any columns with missing values above 50%
        if not missing_values.empty:
            # Create a DataFrame for the missing values
            missing_df = pd.DataFrame({'Column': missing_values.index,
                                       'Missing Values': missing_values.values,
                                       'Missing Values Percentage (%)': ((missing_values / len(data)) * 100).round(2)})
            # Sort the DataFrame by missing value percentages
            missing_df.sort_values(by="Missing Values Percentage (%)", inplace=True)

            # Print the formatted table
            print(missing_df.to_string(index=False))
            print('===' * 26)
        else:
            print("No columns with missing values above 50% found.")
        print()

# Call the function to show missing values in the datasets
show_missing_values(datasets)
```

### Checking for Unique Values

```python
# Define a function to print unique values and missing value percentages for columns with missing values
def print_unique_and_missing(dataset):
    for column in dataset.columns:
        if dataset[column].isnull().any():

            # Calculate the missing values percentage
            missing_percentage = (dataset[column].isnull().sum() / len(dataset)) * 100

            # Create a DataFrame for the unique values and missing value percentage
            data = {
                'Column': [column],
                'Unique Values': [dataset[column].unique()],
                'Missing Values Percentage (%)': [f"{missing_percentage:.2f}%"]
            }
            df = pd.DataFrame(data)

            # Print the DataFrame
            print(df.to_string(index=False))
            print('===' * 26)
            print()


# Print unique values and missing value percentages for columns with missing values in the 'train' dataset
print("Unique values and missing value percentages in the 'train' dataset:")
print_unique_and_missing(train_df)
```

```python
def draw_distplot(column, title=""):
    fig = plt.figure(figsize=(20, 8))
    ax = fig.add_subplot(111)
    ax.set_facecolor((0, 0, 0.10))
    sns.displot(column, kde=True)  # Using displot instead of distplot

    mean_value = column.mean()
    plt.axvline(mean_value, c='red', linewidth=1.5, label="mean")

    median_value = column.median()
    plt.axvline(median_value, c='blue', linewidth=1.5, label="median")
    plt.legend(loc='upper right', prop={'size': 20})
    plt.title(title)
    plt.show()

draw_distplot(column=train_df['age'], title="Age")
```

<h1>Hypothesis Testing and Conclusions</h1>

<p>To test the hypotheses, we'll perform a hypothetical statistical test using a chi-square test to examine the association between educational attainment and income levels.</p>

<strong>Null Hypothesis (H0):</strong>
<p>There is no significant association between educational attainment and income levels.</p>

<strong>Alternate Hypothesis (H1):</strong>
<p>There is a significant association between educational attainment and income levels.</p>

<strong>Here's how we'll proceed:</strong>

<ul>
  <li><strong>Set Significance Level:</strong> Choose a significance level (alpha) to determine the threshold for rejecting the null hypothesis. A common value for alpha is 0.05.</li>
  <li><strong>Perform the Chi-square Test:</strong> Calculate the chi-square statistic and p-value to assess the strength of the association between educational attainment and income levels.</li>
  <li><strong>Interpret the Results:</strong> Compare the p-value to the significance level. If the p-value is less than the significance level, we reject the null hypothesis in favor of the alternative hypothesis, indicating a significant association between educational attainment and income levels. Otherwise, we fail to reject the null hypothesis.</li>
</ul>

<h2><strong>Let's conduct the chi-square test to draw conclusions on these hypotheses.</strong></h2>


```python
# Create contingency table for education and income level
education_income_contingency = pd.crosstab(train_df['education'], train_df['income_above_limit'])

# Perform chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(education_income_contingency)

# Set significance level
alpha = 0.05

# Print results
print("Chi-square Test Results for Education and Income Level:")
print("Chi-square Statistic:", chi2_stat)
print("P-Value:", p_value)

# Compare p-value to significance level
if p_value < alpha:
    print("Conclusion: Reject Null Hypothesis (H0)")
    print("There is a significant association between educational attainment and income levels.")
else:
    print("Conclusion: Fail to Reject Null Hypothesis (H0)")
    print("There is no significant association between educational attainment and income levels.")
```

  <li><strong>Conclusion:</strong> Since the p-value is less than the chosen significance level (alpha = 0.05), we reject the null hypothesis (H0) and conclude that there is a significant association between educational attainment and income levels.</li>
